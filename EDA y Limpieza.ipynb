### Valores ausentes al inicio

CÃ¡lculo valores faltantes:
1. Repetir Ãºltimos valores.
2. Ajuste cuadrÃ¡tico.
3. Multiplicar Ãºltimo valor por 0.95 o 1.05 segÃºn corresponda.

```python
def plot_country(country_code):
    not_null = df[df['Country Code'] == country_code]['GDP'].notnull()
    X = df[df['Country Code'] == country_code]['Year'][not_null].values
    y = df[df['Country Code'] == country_code]['GDP'][not_null].values

    # interpolate nulls
    _y = df[df['Country Code'] == country_code]['GDP'].interpolate(limit_direction='both', order=5).values

    # fit polynomial
    a,b = np.polyfit((X - 1959) ** 2, y, 1)

    # interpolate nulls
    inter_y = df[df['Country Code'] == country_code]['GDP'].interpolate(method='polynomial', order=5)
    back_y = inter_y.interpolate(limit_direction='backward', order=5)
    forw_y = inter_y.interpolate(limit_direction='forward', order=5)

    # backward values
    is_null = forw_y.isnull()
    inter_y[is_null] = 0.95 ** np.array(range(is_null.sum(), 0, -1)) * y[0]

    # forward values
    is_null = back_y.isnull()
    inter_y[is_null] = 1.05 ** np.array(range(is_null.sum())) * y[-1]

    _y2 = inter_y.values

    _x = np.array(range(1960, 2022))
    plt.bar(X, y)
    plt.plot(_x, _y, color='blue', linewidth=3, label="repeat last value")
    plt.plot(_x, _y2, color='red', linewidth=3, label="custom estimation")
    plt.plot(_x, a * (_x - 1959) ** 2 + b, color='green', linewidth=3, label="quadratic fit")
    plt.legend(loc="upper left")
```

La custom-estimation es la que ofrece mejor resultado. AdemÃ¡s de conseguir una curva suave, evitamos valores negativos.

```python
for country_code in df['Country Code'].unique():
    y = df[df['Country Code'] == country_code]['GDP'].dropna().values

    # interpolate nulls
    inter_y = df[df['Country Code'] == country_code]['GDP'].interpolate(method='polynomial', order=5)
    back_y = inter_y.interpolate(limit_direction='backward', order=5)
    forw_y = inter_y.interpolate(limit_direction='forward', order=5)

    # backward values
    is_null = forw_y.isnull()
    inter_y[is_null] = 0.95 ** np.array(range(is_null.sum(), 0, -1)) * y[0]

    # forward values
    is_null = back_y.isnull()
    inter_y[is_null] = 1.05 ** np.array(range(1, is_null.sum() + 1, 1)) * y[-1]

    df.loc[df['Country Code'] == country_code, 'GDP'] = inter_y.values
```
